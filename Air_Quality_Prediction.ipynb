## **Handling the Missing Value**

We will be replacing all the missing values with the mean of that particular feature

1. Convert all the -200 with NaN
2. Replace all NaN values with mean of the specific column

```python
air_quality_data = air_quality_data.replace(to_replace=-200, value=np.NaN)
```

```python
air_quality_data.isnull().sum()
```

```python
air_quality_data.tail()
```

```python
air_quality_data.mean()
```

```python
# Replaceing the missing values with mean values
air_quality_data = air_quality_data.fillna(air_quality_data.mean())
```

```python
air_quality_data.tail()
```

```python
air_quality_data.isnull().sum()
```

Two tasks can be performed on this dataset:
1. Forecasting -> used when we have date and time data in our dataset
2. Regression -> when we have to predict a fixed value and no datetime data given in the dataset

Using Forecasting with **FB Prophet ALgorithm**

1. Converting the Date and Time Column in the dataframe to the FB Prophet format that is YYYY-MM-DD HH:MM:SS

```python
date_info = pd.to_datetime(air_quality_data['Date'])
print(date_info)
```

2. Converting Time column from HH.MM.SS into HH:MM:SS format

```python
time_info = air_quality_data['Time'].apply(lambda x: x.replace('.',':'))
```

```python
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
# combining two series to a pandas dataframe
date_time = pd.concat([date_info, time_info], axis=1)
```

```python
date_time.head()
```

```python
date_time.shape
```

Now we have to convert these two columns into single ds column

```python
# combining the columns
date_time['ds'] = date_time['Date'].astype(str)+' '+date_time['Time'].astype(str)
```

```python
date_time.head()
```

```python
date_time.info()
```

We have to convert ds from object datatype to datetime datatype

```python
data = pd.DataFrame()
```

```python
data['ds'] = pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data.info()
```

```python
data['y'] = air_quality_data['RH']
```

```python
data.head()
```

Importing Prophet

```python
from prophet import Prophet
```

```python
# training the prophet model

model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```
